```python
import requests
from bs4 import BeautifulSoup

def get_bse_sensex():
    url = "https://www.google.com/finance/quote/SENSEX:INDEXBOM"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the element containing the Sensex value
    sensex_value = soup.find('div', class_='YMlKec fxKbKc').text
    return sensex_value

sensex = get_bse_sensex()


def get_bse_stock_price(stock_symbol):
    url = f"https://www.google.com/finance/quote/{stock_symbol}:NSE"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the element containing the stock value
    stock_value_element = soup.find('div', class_='YMlKec fxKbKc')
    if stock_value_element:
        stock_value = stock_value_element.text.replace('₹', '').replace(',', '')
    else:
        stock_value = "N/A"
    return stock_value

def get_previous_close(stock_symbol):
    url = f"https://www.google.com/finance/quote/{stock_symbol}:NSE"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the element containing the previous close value
    previous_close_element = soup.find('div', class_='P6K39c')
    if previous_close_element:
        previous_close = previous_close_element.text.replace('₹', '').replace(',', '')
    else:
        previous_close = "N/A"
    return previous_close
percentage_changes = {}
for stock_name, stock_symbol in stocks.items():
    current_price = float(get_bse_stock_price(stock_symbol))
    previous_close = float(get_previous_close(stock_symbol))
    percentage_change = ((current_price - previous_close) / previous_close) * 100
    percentage_changes[stock_name] = percentage_change


stocks = {
    'Pfizer': 'PFIZER',
    'Ultratech Cement': 'ULTRACEMCO',
    'Reliance Industries': 'RELIANCE',
    'Aditya Birla Capital': 'ABCAPITAL',
    'Grasim Industries': 'GRASIM',
    'Bombay Dyeing': 'BOMDYEING',
    'Raymond': 'RAYMOND',
    'LICI': 'LICI',
    'Tata Steel': 'TATASTEEL',
    'Mahindra Mahindra': 'M%26M',
    'Jio Financial': 'JIOFIN',
    'Raymond Lifestyle': 'RAYMONDLSL'
}

k=[50,15,144,133,95,60,28,15,335,48,72,22]
r=[]
for i in stocks:
    p = float(get_bse_stock_price(stocks[i]))
    r.append(p)

l=[]
for i in range(len(r)):
    l.append(k[i]*r[i])

totinv=sum(l)

```
